## DataFrames
Os DataFrames são conjuntos de dados que tem uma estrutura parecida com a de uma tabela ou planilha do excel, contendo linhas e colunas (*rows, columns*)


```python
dados = {
    "Nome": ["Silas", "Felipe", "Carlos"],
    "Idade": [25,47,32],
}
dataframe = pd.DataFrame(dados) # Cria um DataFrame com base nos dados criados anteriormente
dataframe
```

## Series

```python
s = pd.Series([10,20,30])
s
```

## Series Booleanas
Utilizando operadores lógicos, é possível retornar um série booleana onde o valor da Series será "False" ou "True", baseado no retorno da operação de cada elemento

```python
df['number of reviews'] > 30
```

## Filtros
Utilizando essa mesma ideia, é possível utilizar filtros em um Dataframe para que ele retorne **APENAS** os que satisfazem as condições dadas pelos filtros

```python
# Criando Filtros
reviews_greater_than_30 = df['number of reviews'] > 30
host_identity_verified = df['host_identity_verified'] == 'verified'

# Imprimindo o dataframe ambos os filtros aplicados
df[reviews_greater_than_30 & host_identity_verified]
```

## Criando e removendo colunas
É possível também criar novas colunas e inseri-las no dataframe,
assim como também é possível remover.

```python
df['Open for 6 months'] = df['availability 365'] > 182 # Coluna com apenas as casas disponiveis por mais da metade do ano
df
```

```python
df = df.drop('Open for 6 months', axis=1) # O metodo drop pode remover linhas ou colunas, para especificar que uma coluna será removida, é preciso especificar o axis=1
df
```

## Resumindo dados

## Criando estatísticas
Esses são alguns exemplos de funções que leem os dados e com base neles retornam um valor de acordo com a operação feita.

```python
df['number of reviews'].sum() # Retorna a soma de todos os elementos de uma coluna
```

```python
df['number of reviews'].mean() # Retorna a média de uma coluna
```

## Limpeza de dados
Nem sempre os dados terão todas as informações necessárias, as vezes, algumas linhas terão valores nulos, por isso, é preciso fazer uma limpeza de dados para saber quais devem ser removidos ou substituidos com um valor médio

```python
df.dropna() # Remove todas as linhas com algum valor nulo na tabela
```

```python
df.dropna(subset='availability 365') # Remove todas as linhas que possuem valor nulo na coluna availabilty 365
```

```python
media_availability365 = df['availability 365'].mean() # Tira uma média da coluna availability 365
df['availability 365'] = df['availability 365'].fillna(media_availability365) # Preenche os valores nulos da coluna com a media
df
```

```python
df['availability 365'] = df['availability 365'].astype(int) # Converte os valores da coluna para int
```

## Contando ocorrências
É possível contar ocorrências de um valor na tabela utilizando a função **value_counts()**

```python
df['host_identity_verified'].value_counts() # Retorna o total de anfitrioes confirmados e os nao confirmados
```

```python
df['neighbourhood'].value_counts() # Retorna o total de airbnb em cada bairro
```

# Operações de agregação
Quando for preciso, é possível calcular uma métrica baseada em uma coluna utilizando o **groupby()**


```python
df.groupby('neighbourhood')['number of reviews'].mean() # Retorna uma media do numero de reviews por bairro
```

```python
df.groupby('neighbourhood')['number of reviews'].sum() # Retorna uma soma de todas as reviews por bairro
```

```python
df.groupby('neighbourhood')['number of reviews'].max() # Retorna o maior numero de reviews no bairro
```
